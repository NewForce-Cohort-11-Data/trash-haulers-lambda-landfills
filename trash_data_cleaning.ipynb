```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
# Read in csv and get dataframe info
trash_df = pd.read_csv(r'../trash-haulers-lambda-landfills/trash_hauler_report.csv')
print(trash_df.shape)
trash_df
```

```python
# Check column types and missing values
trash_df.info()
# Basic statistics for numeric columns
trash_df.describe()
```

```python
# Count missing values per column
trash_df.isnull().sum()
```

```python
# Replace spaces with underscores and convert to lowercase
trash_df.columns = trash_df.columns.str.replace(" ", "_").str.title()
trash_df
```

```python
# Convert date column to datetime
trash_df['Date_Opened'] = pd.to_datetime(trash_df['Date_Opened'], errors='coerce')
trash_df
```

```python
# Filter for missed pickups
missed_trash_df = trash_df[
    (trash_df['Request_'].str.contains("missed", case=False, na=False)) |
    (trash_df['Description'].str.contains("miss", case=False, na=False))
].copy()
```

```python
# Count missed pickups
missed_counts = missed_trash_df.groupby('Incident_Address').size().reset_index(name='missed_count')
missed_counts
```

```python
# Count missed > 1
missed_counts['fine'] = (missed_counts['missed_count'] - 1).clip(lower=0) * 200
```

```python
# Calculate total fines
total_fines = missed_counts['fine'].sum()

print("Total fines due: $", total_fines)
```
